```python
import pandas as pd
import matplotlib.pyplot as plt
play = pd.read_csv('play.csv')
```

```python
grouped = play.groupby(play.columns[7])
counts = grouped.size().sort_values(ascending=False)
for value, count in counts.items():
    print(f'{value}: {count}')
```

```python
counts = play['neighborhood'].value_counts()

plt.figure(figsize=(16, 6))

counts.plot(kind='bar')

plt.xlabel('Neighborhood')
plt.ylabel('Amount Of Parks')
plt.title('Amount Of Parks Per Neighborhood')

plt.show()
```
