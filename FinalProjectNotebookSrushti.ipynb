## The submetric I picked for the safest neighborhood in Pittsburgh is the enrollment of kids at schools in different neighborhoods

School enrollment in an area can be an important safety measure to consider because it can help create a stronger engaging community, which can lead to a reduction in crime and safety of the environment over time. Some measure to consider are: 
1. Sense of community: Schools can help build a sense of community and encourage residents to participate in more activities in their neighborhoods.
2. It grants access to resources: Schools often provide access to resources that can help improve safety in an area.
3. Education and other opportunities: Access ot education is linked to better employment in the future for residents, which can help reduce poverty and crime in the area** 

Here I pulled out the data from the dataset "School Enrollment, All Races" and made a csv file of the names of the neighborhood and the enrollement total ranking from the highest to the lowest for the top 30 neighborhoods. 

```python
import pandas as pd
import matplotlib.pyplot as plt
df1 = pd.read_csv('e.csv')
df2 = pd.read_csv('enroll.csv')
```

```python
df1['Enrollment Total'] = df1['Estimate; Enrolled in school:'].astype(int)

top_30 = df1[['Neighborhood', 'Enrollment Total']].sort_values(by='Enrollment Total', ascending=False).head(30)
print(top_30)
```

**Explanation: The first part reads out two csv files into two separate data frames df1 and df2. The second part creates a new column called 'Enrollment Total' in 'df1' which contains the integer values of the 'Estimate; Enrolled in school:' column. Then, it sorts the 'Neighborhood' column in descending order based on the 'Enrollment Total' column values, and selectes the top 30 rows with the highest enrollement totals. Finally, the 'print' function dispays everything.**

I even graphed it so it was clear to see which neighborhood had the highest enrollement total. 

```python
enrollment_by_neighborhood = df1.groupby('Neighborhood')['Estimate; Enrolled in school:'].sum()
enrollment_by_neighborhood.plot(kind='bar', figsize=(12, 6))

plt.title('Enrollment in School by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Students Enrolled')

plt.show()
```

**Explanation: The first line of code groups the data in the df1 DataFrame by neighborhood using the groupby() function and then calculates the sum of the Estimate; Enrolled in school: column for each neighborhood. Secondly, it creates a bar plot using the 'plot' function with the 'kind' parameter set to 'bar' to give us a bar chart. The 'figsize' parameter specifies the dimensions of the plot, with a width of 12 adn height of 6. Lastly, the code sets the title of the plot to 'Enrollment in School by Neighborhood' and the 'xlabel' and 'ylabel' paramters are used to label the x and y axis. Finally, the 'show' function prints out the graph for us**

According to the graph, the safest neighborhood based on the enrollement total at schools is North Oakland because it has a total of 7192 students annually. Partially the reason why the enrollement rate is high at this neighborhood is because it is home to universities like the University of Pittsburgh and Carnegie Mellon University. These universities attract a large number of students, faculty and staff to the area, which can contribute to a higher enrollement total at schools.

## The above dataset showed the data for the enrollement of students of all races in the past. I wanted a data that was somewhat current. I found one that gave me the enrollment by school and neighborhood 2020-2021 school year. Here are the results. 

```python
df2 = df2[['neighborhood','enrolled_students']]
df2 = df2.groupby(['neighborhood', 'enrolled_students']).sum().reset_index()
df2 = df2.sort_values(by='enrolled_students', ascending=False).head(30)
df2.to_csv('enrollment_top_30.csv', index=False) 

df3 = pd.read_csv('enrollment_top_30.csv')

print(df3)
```

**Explanation: The first line of the code selects only the 'neighborhood' and 'enrolled_students' columns from 'df2' and then groups the data along with calculating the sum of each group. Next, 'df2' is sorted in descending order based on the 'enrolled_students' values and then selects the top 30 neighborhoods with the highest enrollement. Finally everything is printed and displayed.**

## Here is the graph for the dataset above. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the enrollment data
df3 = pd.read_csv('enrollment_top_30.csv')

# Create a horizontal bar chart
plt.barh(df3['neighborhood'], df3['enrolled_students'], color='green')

# Set the chart title and axis labels
plt.title('Top 30 Neighborhoods by Enrollment')
plt.xlabel('Enrolled Students')
plt.ylabel('Neighborhood')

# Show the chart
plt.show()
```

**Explanation: Firstly, 'enrollment_top_30.csv' file is read. Next, a horizontal bar chart is created using the 'plt.barh' function from the matplotlib library. The first part is the data for the x-axis which is 'enrolled_students' column of 'df3' and the second part is the data for the y-axis which is the 'neighborhood' column of 'df3'. I made it a different color using the 'color' function just to add a little creativity. The chart title, x-axis label and y-axis label are set using the 'plt.title', 'plt.xlabel' and 'plt.ylabel' functions. Finally, everything is displayed using the 'plt.show' function.**
