**First, I import pandas, as well as the csv files I will be using as my submetric.**

```python
import pandas as pd
crime = pd.read_csv('crime.csv')
pop = pd.read_csv('pops.csv')
```

**I started by printing out the neighborhoods in order of most reported crimes, to least.**

```python
grouped = crime.groupby(crime.columns[6])
counts = grouped.size().sort_values(ascending=False)
for value, count in counts.items():
    print(f'{value}: {count}')
```

**Now that we have the Population data in a csv and a list of the numbers we want to divide that column by, we can calculate the crime rate per capita in Pitssburgh neighborhoods:**

```python
pop['2020_Total_Population'] = pop['2020_Total_Population'].astype(int)
perCapita = pop['2020_Total_Population'].div(pd.Series(numbers)).astype(int)
print(perCapita)
```
